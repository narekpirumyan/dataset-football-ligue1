## Nombre de matchs par jour de match

Compte le nombre de matchs pour chaque matchday (1–38) à partir des résultats de match.

```python
import pandas as pd

# Charger les résultats de match (une ligne = un match)
# Chemin depuis le dossier du notebook (Dataset/)
df_matches = pd.read_csv("match_results_2023_2024.csv")

# Nombre de matchs par jour de match
matches_per_matchday = df_matches.groupby("matchday").size().sort_index()
matches_per_matchday.name = "nombre_matchs"

print("Matchs par jour de match (1–38):")
print(matches_per_matchday.to_string())

# Résumé
print(f"\nTotal matchdays: {len(matches_per_matchday)}")
print(f"Total matchs: {matches_per_matchday.sum()}")
print(f"Min / Max matchs par jour: {matches_per_matchday.min()} / {matches_per_matchday.max()}")
```

## Nombre de buts : Enzo Ndombele et Florian Morel

Graphique en barres du nombre de buts (saison) pour ces deux joueurs, à partir de `player_stats_season.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_players = pd.read_csv("player_stats_season.csv")
selected = df_players[df_players["full_name"].isin(["Enzo Ndombele", "Florian Morel"])][["full_name", "goals", "club", "position"]]

fig, ax = plt.subplots(figsize=(6, 3.5))
bars = ax.bar(selected["full_name"], selected["goals"], color=["#1f77b4", "#ff7f0e"], edgecolor="black", linewidth=0.8)
ax.set_ylabel("Nombre de buts")
ax.set_title("Buts marqués — Enzo Ndombele & Florian Morel (saison 2023-2024)")
ax.set_ylim(0, max(selected["goals"]) * 1.2 if selected["goals"].max() > 0 else 2)
for bar, val in zip(bars, selected["goals"]):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, str(int(val)), ha="center", va="bottom", fontsize=12)
plt.tight_layout()
plt.show()
```

## Nombre de buts par attaquant (Forward)

Graphique du nombre de buts de chaque joueur à la position **Forward** (attaquant), à partir de `player_stats_season.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_players = pd.read_csv("player_stats_season.csv")
attackers = df_players[df_players["position"] == "Forward"][["full_name", "goals", "club"]].sort_values("goals", ascending=True)

fig, ax = plt.subplots(figsize=(8, max(6, len(attackers) * 0.35)))
bars = ax.barh(attackers["full_name"], attackers["goals"], color="steelblue", edgecolor="black", linewidth=0.6)
ax.set_xlabel("Nombre de buts")
ax.set_title("Buts marqués par attaquant — saison 2023-2024")
ax.set_xlim(0, attackers["goals"].max() * 1.15 if len(attackers) else 1)
for bar, val in zip(bars, attackers["goals"]):
    ax.text(val + 0.15, bar.get_y() + bar.get_height() / 2, str(int(val)), va="center", fontsize=9)
plt.tight_layout()
plt.show()
```
